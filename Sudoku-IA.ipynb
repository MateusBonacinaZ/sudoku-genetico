# IMPORTAÇÃO DE BIBLIOTECAS

```python
# IMPORTAÇÃO DAS BIBLIOTECAS
from random import randint, choice, uniform
import copy
```

# CRIAÇÃO E EXIBIÇÃO DO TABULEIRO

```python
# FUNÇÃO PARA CRIAÇÃO DO TABULEIRO SEGUINTO O TEMPLATE PRÉ-DEFINIDO ACIMA
def create_board()-> list:
    matrix = [
                ['_', 3, 4, 6, 7, 2, 1, 9, 8],
                [6, 7, 8, 1, 9, 5, 3, 4, 2],
                [9, 1, 2, 3, 4, 8, 5, 6, 7],
                [8, 5, 9, 4, '_', 6, 7, 1, 3],
                [7, 6, 1, 8, 5, 3, 9, 2, 4],
                [4, '_', 3, 7, 9, 1, '_', 5, 6],
                [9, 6, 1, '_', 8, 7, 3, 4, 5],
                [5, 3, 7, 4, 1, 9, 2, 8, 6],
                [2, 8, 4, 6, 3, 5, 1, 7, 9]
            ]
    
    return matrix
```

```python
# FUNÇÃO PARA COMPLETAR OS ESPAÇOS EM BRANCO COM NÚMEROS ALEATÓRIOS
def fill_in_spaces(board: list, spaces: list) -> list:
    for space in spaces:
        board[space[0]][space[1]] = randint(1, 9)
    return board
```

```python
# FUNÇÃO PARA EXIBIR O TABULEIRO DE FORMA CLARA PARA O USUÁRIO
def show_board(board: list):
    print(f"|{board[0][0]} {board[0][1]} {board[0][2]}| |{board[1][0]} {board[1][1]} {board[1][2]}| |{board[2][0]} {board[2][1]} {board[2][2]}|")
    print(f"|{board[0][3]} {board[0][4]} {board[0][5]}| |{board[1][3]} {board[1][4]} {board[1][5]}| |{board[2][3]} {board[2][4]} {board[2][5]}|")
    print(f"|{board[0][6]} {board[0][7]} {board[0][8]}| |{board[1][6]} {board[1][7]} {board[1][8]}| |{board[2][6]} {board[2][7]} {board[2][8]}|", end='\n\n')
    
    print(f"|{board[3][0]} {board[3][1]} {board[3][2]}| |{board[4][0]} {board[4][1]} {board[4][2]}| |{board[5][0]} {board[5][1]} {board[5][2]}|")
    print(f"|{board[3][3]} {board[3][4]} {board[3][5]}| |{board[4][3]} {board[4][4]} {board[4][5]}| |{board[5][3]} {board[5][4]} {board[5][5]}|")
    print(f"|{board[3][6]} {board[3][7]} {board[3][8]}| |{board[4][6]} {board[4][7]} {board[4][8]}| |{board[5][6]} {board[5][7]} {board[5][8]}|", end='\n\n')
    
    print(f"|{board[6][0]} {board[6][1]} {board[6][2]}| |{board[7][0]} {board[7][1]} {board[7][2]}| |{board[8][0]} {board[8][1]} {board[8][2]}|")
    print(f"|{board[6][3]} {board[6][4]} {board[6][5]}| |{board[7][3]} {board[7][4]} {board[7][5]}| |{board[8][3]} {board[8][4]} {board[8][5]}|")
    print(f"|{board[6][6]} {board[6][7]} {board[6][8]}| |{board[7][6]} {board[7][7]} {board[7][8]}| |{board[8][6]} {board[8][7]} {board[8][8]}|", end='\n\n')
```

# FUNÇÕES AUXILIARES (LOCALIZAR ESPAÇOS EM BRANCO, CALCULAR CONFLITOS...)

```python
# FUNÇÃO PARA LOCALIZAR OS ESPAÇOS EM BRANCO
def find_spaces(state: list) -> list:
    spaces = []
    for submatrix in range(len(state)):
        for cell in range(len(state[submatrix])):
            if state[submatrix][cell] == '_':
                spaces.append((submatrix, cell)) 
    return spaces
```

```python
# FUNÇÃO PARA CALCULAR A QUANTIDADE DE CONFLITOS NO TABULEIRO
def calculate_conflicts(state: list) -> int:
    conflicts = 0

    # VERFICIANDO CONFLITOS DENTRO DAS SUBMATRIZES
    for submatrix in range(len(state)):
        numbers_conflicted_submatrix = []
        for cell in range(len(state[submatrix])):
            number = state[submatrix][cell]
            if number not in numbers_conflicted_submatrix:
                quantity_conflicted_submatrix = state[submatrix].count(number)
                if quantity_conflicted_submatrix > 1:
                    conflicts += quantity_conflicted_submatrix-1
                    numbers_conflicted_submatrix.append(number)

    # VERIFICANDO CONFLITOS NA LINHA
    lines = converct_lines(state)
    for line in range(len(lines)):
        numbers_conflicted_lines = []
        for cell in range(len(lines[line])):
            number = lines[line][cell]
            if number not in numbers_conflicted_lines:
                quantity_conflicted_line = lines[line].count(number)
                if quantity_conflicted_line > 1:
                    conflicts += quantity_conflicted_line-1
                    numbers_conflicted_lines.append(number)

    # VERIFICANDO CONFLITOS NA COLUNA
    columns = converct_columns(state)
    for column in range(len(columns)):
        numbers_conflicted_columns = []
        for cell in range(len(columns[column])):
            number = columns[column][cell]
            if number not in numbers_conflicted_columns:
                quantity_conflicted_column = columns[column].count(number)
                if quantity_conflicted_column > 1:
                    conflicts += quantity_conflicted_column-1
                    numbers_conflicted_columns.append(number)
    
    return conflicts
```

```python
# FUNÇÃO PARA CRIAR AS LINHAS DO TABULEIRO PARA VERIFICAÇÃO DE CONFLITO
def converct_lines(state: list) -> list:
    lines = []

    lines.append(state[0][0:3]+state[1][0:3]+state[2][0:3])
    lines.append(state[0][3:6]+state[1][3:6]+state[2][3:6])
    lines.append(state[0][6:9]+state[1][6:9]+state[2][6:9])

    lines.append(state[3][0:3]+state[4][0:3]+state[5][0:3])
    lines.append(state[3][3:6]+state[4][3:6]+state[5][3:6])
    lines.append(state[3][6:9]+state[4][6:9]+state[5][6:9])

    lines.append(state[6][0:3]+state[7][0:3]+state[8][0:3])
    lines.append(state[6][3:6]+state[7][3:6]+state[8][3:6])
    lines.append(state[6][6:9]+state[7][6:9]+state[8][6:9])
    
    return lines
```

```python
# FUNÇÃO PARA CRIAR AS COLUNAS DO TABULEIRO PARA VERIFICAÇÃO DE CONFLITO
def converct_columns(state: list) -> list:
    columns = []
    columns.append([state[0][0], state[0][3], state[0][6], state[3][0], state[3][3], state[3][6], state[6][0], state[6][3], state[6][6]])
    columns.append([state[0][1], state[0][4], state[0][7], state[3][1], state[3][4], state[3][7], state[6][1], state[6][4], state[6][7]])
    columns.append([state[0][2], state[0][5], state[0][8], state[3][2], state[3][5], state[3][8], state[6][2], state[6][5], state[6][8]])

    columns.append([state[1][0], state[1][3], state[1][6], state[4][0], state[4][3], state[4][6], state[7][0], state[7][3], state[7][6]])
    columns.append([state[1][1], state[1][4], state[1][7], state[4][1], state[4][4], state[4][7], state[7][1], state[7][4], state[7][7]])
    columns.append([state[1][2], state[1][5], state[1][8], state[4][2], state[4][5], state[4][8], state[7][2], state[7][5], state[7][8]])

    columns.append([state[2][0], state[2][3], state[2][6], state[5][0], state[5][3], state[5][6], state[8][0], state[8][3], state[8][6]])
    columns.append([state[2][1], state[2][4], state[2][7], state[5][1], state[5][4], state[5][7], state[8][1], state[8][4], state[8][7]])
    columns.append([state[2][2], state[2][5], state[2][8], state[5][2], state[5][5], state[5][8], state[8][2], state[8][5], state[8][8]])
    
    return columns
```

# FUNÇÕES DO ALGORITMO GENÉTICO

```python
# FUNÇÃO PARA GERAR A POPULAÇÃO INICIAL
def generate_initial_population(population_size: int, empty_spaces: list) -> list:
    population = []
    for individual in range(population_size):
        board = create_board()
        board = fill_in_spaces(board, empty_spaces)
        population.append(board)
    return population
```

```python
# FUNÇÃO DE CROSS OVER (CRUZAMENTO) DE INDIVÍDUOS
def cross_over(genome_1: list, genome_2: list) -> list:
    g_1 = copy.deepcopy(genome_1)
    g_2 = copy.deepcopy(genome_2)
    cut = randint(0, len(g_1)-1)
    return g_1[:cut] + g_2[cut:], g_2[:cut] + g_1[cut:]
```

```python
# FUNÇÃO DE MUTAÇÃO QUE ALTERA UMA QUANTIDADE ESPECÍFICA DE ELEMENTOS DA SUBMATRIZ
def mutation(state: list, quantity_mutations: int, empty_spaces: list) -> list:
    empty_spaces_local = copy.deepcopy(empty_spaces)
    for quantity in range(quantity_mutations):
        space = choice(empty_spaces_local)
        empty_spaces_local.remove(space)
        number = randint(1, 9)
        while number == state[space[0]][space[1]]:
            number = randint(1, 9)
        state[space[0]][space[1]] = number
    return state
```

```python
# FUNÇÃO FITNESS PARA AVALIAR A POPULAÇÃO
def fitness(population: list) -> list:
    fitness_values = []
    for individual in population:
        fitness_values.append(1 - calculate_conflicts(individual)/216)
    return fitness_values
```

```python
# FUNÇÃO DE ROLETA VICIADA PARA SELECIONAR INDIVÍDUOS
def addicted_roulette(population: list, fitness: list) -> list:
    fitness_total = sum(fitness)
    fractions = [f/fitness_total for f in fitness]
    random_number = uniform(0, 1)
    acum = 0
    for i, individual in enumerate(population):
        acum += fractions[i]
        if acum >= random_number:
            return individual
    
    return choice(population)
```

```python
# FUNÇÃO UTILIZADA PARA SELECIONAR UMA DETERMINADA QUANTIDADE DE CASAIS REPRODUTORES
def natural_selection(population: list, number_breeding_couples: int) -> list:
    selection = []
    fitness_value = fitness(population)
    for individual in range(2*number_breeding_couples):
        selection.append(addicted_roulette(population, fitness_value))
    return selection
```

```python
# FUNÇÃO PARA VERIFICAR SE A META FOI ALCANÇADA
def meta_test(population: list):
    fitness_value = fitness(population)
    try:
        position_meta = fitness_value.index(1)
        return population[position_meta]
    except (ValueError):
        return -1
```

```python
# FUNÇÃO PARA LOCALIZAR O MELHOR INDIVÍDUO DA GERAÇÃO
def best_individual(population: list) -> list:
    fitness_value = fitness(population)
    position = fitness_value.index(max(fitness_value))
    return (population[position], fitness_value[position])
```

```python
# FUNÇÃO DO ALGORITMO GENÉTICO
def genetic(population_size: int, generations: int, empty_spaces: list, quantity_mutations: int, mutation_percentage: int, number_breeding_couples: int):
    
    population = generate_initial_population(population_size, empty_spaces)
    best = best_individual(population)
    print(f"MELHOR ÍNDIVIDUO DA GERAÇÃO 0 COM FITNESS: {best[1]}")
    
    for generation in range(generations):
        print(f"GERAÇÃO: {generation}")
        
        meta = meta_test(population)
        if meta != -1:
            print(f"\nSOLUÇÃO ENCONTRADA NA GERAÇÃO: {generation}\n")
            return show_board(meta)
        else:
            best_geration = best_individual(population)
            if best_geration[1] > best[1]:
                best = best_geration
                print(f'NOVO MELHOR FITNESS COM VALOR: {best[1]}')
            

        new_population = []
        breeders = natural_selection(population, number_breeding_couples)
        males = breeders[:number_breeding_couples]
        females = breeders[number_breeding_couples:]
        for male, female in zip(males, females):
            son_1, son_2 = cross_over(male, female)
            new_population.append(son_1)
            new_population.append(son_2)

        meta = meta_test(new_population)
        if meta != -1:
            print(f"\nSOLUÇÃO ENCONTRADA NA GERAÇÃO: {generation}\n")
            return show_board(meta)
        else:
            best_geration = best_individual(new_population)
            if best_geration[1] > best[1]:
                best = best_geration
                print(f'NOVO MELHOR FITNESS COM VALOR: {best[1]}')

        for individual in new_population:
            if randint(0, 100) < mutation_percentage:
                individual = mutation(individual, quantity_mutations, empty_spaces)

        population = new_population

    print(f"\nALGORITMO ENCERRADO DEVIDO AO NÚMERO DE GERAÇÕES SEM ENCONTRAR UM RESULTADO SATISFATÓRIO.\nO MELHOR INDIVÍDUO OBTEVE O FITNESS DE {best[1]}.\n\n")
    return show_board(best[0])
```

# CONFIGURAÇÕES E INICIALIZAÇÃO DO ALGORITMO GENÉTICO

```python
# CONFIGURAÇÕES INICIAIS DO ALGORITMO GENÉTICO
population_size = 1000
generations = 10000
empty_spaces = find_spaces(create_board())

quantity_mutations = 2
if quantity_mutations > len(empty_spaces):
    raise ValueError("O NÚMERO DE VALORES ALTERADO POR MUTAÇÃO É MAIOR QUE OS ESPAÇOS VAZIOS!")

mutation_percentage = 50
if mutation_percentage > 100:
    raise ValueError("A PORCENTAGEM DE CHANCE DE MUTAÇÃO É MAIOR QUE 100%.")

number_breeding_couples = 50

# INICIALIZAÇÃO DO ALGORITMO GENÉTICO
genetic(population_size, generations, empty_spaces, quantity_mutations, mutation_percentage, number_breeding_couples)
```
